```
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

image = Image.open('/content/drive/MyDrive/Impulse_19/Task#4/Barbara.png')
image = np.array(image)
print(image.shape)
```

**BILINEAR INTERPOLATION**

```
upscale_img = np.zeros((256, 256))        # upscaled image

for i in range(0, 256, 2):                            # retaining the original image by adding empty columns and rows alternatively
  for j in range(0, 256, 2):
    upscale_img[i, j] = image[int(i/2), int(j/2)]

for i in range(1, 255, 2):                                      #interpolating
  for j in range(1, 255, 2):
    Ix1 = (upscale_img[i-1, j-1] + upscale_img[i-1, j+1])/2
    Ix2 = (upscale_img[i+1, j-1] + upscale_img[i+1, j+1])/2
    I = (Ix1 + Ix2)/2
    upscale_img[i, j] = I

for i in range(1, 255, 2):                                      #interpolating
  for j in range(0, 256, 2):
    Ix1 = (upscale_img[i-1, j] + upscale_img[i+1, j])/2
    Ix2 = (upscale_img[i, j-1] + upscale_img[i, j+1])/2
    I = (Ix1 + Ix2)/2
    upscale_img[i, j] = I

for i in range(0,256, 2):                                      #interpolating
  for j in range(1, 255, 2):  
    Ix1 = (upscale_img[i-1, j] + upscale_img[i+1, j])/2
    Ix2 = (upscale_img[i, j-1] + upscale_img[i, j+1])/2
    I = (Ix1 + Ix2)/2
    upscale_img[i, j] = I

upscale_img[:, -1] = upscale_img[:, -2]
upscale_img[-1, :] = upscale_img[-2, :]
```

```
plt.imshow(image, cmap = 'gray')
plt.show()
plt.imshow(upscale_img, cmap = 'gray')
plt.show()
```
